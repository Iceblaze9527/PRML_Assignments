```python
import pandas as pd
from sklearn.manifold import MDS
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
```

```python
path='HomeworkData/'

data_frame = pd.read_excel(path+'city_dist.xlsx', index_col = 0)
X = data_frame.values
names = data_frame.columns.values
myfont = FontProperties(fname='/System/Library/Fonts/Supplemental/Songti.ttc',size=12)
```

```python
model = MDS(n_components=2, metric=True, random_state=23)
X_mds = model.fit_transform(X)

plt.figure(figsize=(16,12))
plt.scatter(X_mds [:,0], X_mds [:,1], c='r')

for i in range(X_mds.shape[0]):
    plt.annotate(s=names[i], xy=(X_mds[i,0],X_mds[i,1]), fontproperties=myfont)
```
