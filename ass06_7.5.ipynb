```python
from scipy.stats import norm, uniform
import numpy as np
import matplotlib.pyplot as plt
```

## 1. 最大似然估计

```python
x11 = norm.rvs(size=10,random_state=30)
x12 = norm.rvs(size=10,random_state=31)
x13 = norm.rvs(size=10,random_state=1)

x21 = norm.rvs(size=100,random_state=30)
x22 = norm.rvs(size=100,random_state=31)
x23 = norm.rvs(size=100,random_state=1)

x31 = norm.rvs(size=1000,random_state=30)
x32 = norm.rvs(size=1000,random_state=31)
x33 = norm.rvs(size=1000,random_state=1)
```

### 1.1 N=10

```python
mu11, sig11 = norm.fit(x11)
mu12, sig12 = norm.fit(x12)
mu13, sig13 = norm.fit(x13)

print("MLE with N=10:")
print("1. mu11=%.4f sig11=%.4f"%(mu11,sig11))
print("2. mu12=%.4f sig12=%.4f"%(mu12,sig12))
print("3. mu13=%.4f sig13=%.4f"%(mu13,sig13))

x = np.linspace(-5,5,100)
plt.plot(x, norm.pdf(x,loc = mu11,scale = sig11), 'r-', label='x11')
plt.plot(x, norm.pdf(x,loc = mu12,scale = sig12), 'c-', label='x12')
plt.plot(x, norm.pdf(x,loc = mu13,scale = sig13), 'm-', label='x13')
plt.plot(x, norm.pdf(x), 'b-', label='std')
plt.legend()
plt.show()
```

### 1.2 N=100

```python
mu21, sig21 = norm.fit(x21)
mu22, sig22 = norm.fit(x22)
mu23, sig23 = norm.fit(x23)

print("MLE with N=100:")
print("1. mu21=%.4f sig21=%.4f"%(mu21,sig21))
print("2. mu22=%.4f sig22=%.4f"%(mu22,sig22))
print("3. mu23=%.4f sig23=%.4f"%(mu23,sig23))

x = np.linspace(-5,5,100)
plt.plot(x, norm.pdf(x,loc = mu21,scale = sig21), 'r-', label='x21')
plt.plot(x, norm.pdf(x,loc = mu22,scale = sig22), 'c-', label='x22')
plt.plot(x, norm.pdf(x,loc = mu23,scale = sig23), 'm-', label='x23')
plt.plot(x, norm.pdf(x), 'b-', label='std')
plt.legend()
plt.show()
```

### 1.3 N=1000

```python
mu31, sig31 = norm.fit(x31)
mu32, sig32 = norm.fit(x32)
mu33, sig33 = norm.fit(x33)

print("MLE with N=1000:")
print("1. mu31=%.4f sig31=%.4f"%(mu31,sig31))
print("2. mu32=%.4f sig32=%.4f"%(mu32,sig32))
print("3. mu33=%.4f sig33=%.4f"%(mu33,sig33))

x = np.linspace(-5,5,100)
plt.plot(x, norm.pdf(x,loc = mu31,scale = sig31), 'r-', label='x31')
plt.plot(x, norm.pdf(x,loc = mu32,scale = sig32), 'c-', label='x32')
plt.plot(x, norm.pdf(x,loc = mu33,scale = sig33), 'm-', label='x33')
plt.plot(x, norm.pdf(x), 'b-', label='std')
plt.legend()
plt.show()
```

## 2. 贝叶斯估计-均值

```python
def mu_N(x, m_N, N = 1000, mu_0 = -5):
    return N/(N+x)*m_N + x/(N+x)*mu_0

def sig_N(x, N = 1000, sig_sq=1):
    return sig_sq/(N+x)
```

```python
mu1 = mu_N(100, mu31)
sig1 = sig_N(100)

mu2 = mu_N(10, mu31)
sig2 = sig_N(10)

mu3 = mu_N(1, mu31)
sig3 = sig_N(1)

mu4 = mu_N(0.1, mu31)
sig4 = sig_N(0.1)

print("BE with N=1000:")
print("1. mu1=%.4f sig1=%.4f"%(mu1,sig1))
print("2. mu2=%.4f sig2=%.4f"%(mu2,sig2))
print("3. mu3=%.4f sig3=%.4f"%(mu3,sig3))
print("4. mu4=%.4f sig4=%.4f"%(mu4,sig4))

x = np.linspace(-3,3,10000)
plt.plot(x, norm.pdf(x,loc = mu1, scale = 1), 'r-', label='ratio=100')
plt.plot(x, norm.pdf(x,loc = mu2, scale = 1), 'c-', label='ratio=10')
plt.plot(x, norm.pdf(x,loc = mu3, scale = 1), 'm-', label='ratio=1')
plt.plot(x, norm.pdf(x,loc = mu4, scale = 1), 'g-', label='ratio=0.1')
plt.plot(x, norm.pdf(x), 'b-', label='std')
plt.legend()
plt.show()
```

## 3. 一致分布样本的最大似然估计

```python
y1 = uniform.rvs(size=100,random_state=30)
y2 = uniform.rvs(size=100,random_state=31)
y3 = uniform.rvs(size=100,random_state=1)
```

```python
muU1, sigU1 = norm.fit(y1)
muU2, sigU2 = norm.fit(y2)
muU3, sigU3 = norm.fit(y3)

print("MLE with N=100:")
print("1. muU1=%.4f sigU1=%.4f"%(muU1,sigU1))
print("2. muU2=%.4f sigU2=%.4f"%(muU2,sigU2))
print("3. muU3=%.4f sigU3=%.4f"%(muU3,sigU3))

x = np.linspace(-5,5,100)
plt.plot(x, norm.pdf(x,loc = muU1,scale = sigU1), 'r-', label='y1')
plt.plot(x, norm.pdf(x,loc = muU2,scale = sigU2), 'c-', label='y2')
plt.plot(x, norm.pdf(x,loc = muU3,scale = sigU3), 'm-', label='y3')
plt.plot(x, uniform.pdf(x), 'b-', label='std')
plt.legend()
plt.show()
```
