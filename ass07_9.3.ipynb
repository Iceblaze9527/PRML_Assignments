```python
import numpy as np
from hmmlearn.hmm import MultinomialHMM
```

```python
path="HomeworkData/"
seq = np.load(path+"sequences.npy")
```

```python
model = MultinomialHMM(n_components=2,n_iter=1000) 
model.startprob_ = 1/2*np.ones(2)
model.transmat_ = 1/2*np.ones((2,2)) 
model.emissionprob_ = 1/6*np.ones((2,6)) 
model.fit(seq.reshape(-1,1)-1,lengths=(30*np.ones(200)).astype(np.uint8))
```

```python
np.set_printoptions(precision=4,suppress=True)

pi = model.startprob_
A = model.transmat_
E = model.emissionprob_

print(pi)
print(A)
print(E)
```

```python
def foreward(pi,A,e):
    alpha = e*pi
    print("alpha_1:",alpha)
    for i in range(3):
        alpha = e*np.dot(A.T,alpha)
        print("alpha_%d:"%(i+2),alpha)
    prob = np.sum(alpha)
    print("prob:",prob)

def backward(pi,A,e):
    beta = np.ones(2)
    print("beta_4:",beta)
    for i in range(3):
        beta = np.dot(A, e*beta)
        print("beta_%d:"%(3-i),beta)
    prob = np.dot(pi,e*beta)
    print("prob:",prob)
```

```python
e = np.array([E[0,5],E[1,5]])
foreward(pi,A,e)
backward(pi,A,e)
```

```python
obs = (np.array([3,2,1,3,4,5,6,3,1,4,1,6,6,2,6]).reshape(-1,1))-1
log_prob, state_seq = model.decode(obs)
print(state_seq)
```
